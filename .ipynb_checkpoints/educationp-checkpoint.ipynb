# Test and Train

```python
from sklearn.model_selection import train_test_split
# importing train_test_split  from sklearn.model_selection
```

```python
x_train,x_test,y_train, y_test=train_test_split(x,y,test_size=0.3, random_state=43)
# train and test size
```

# scailing

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# import linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

# Model training  Fit Models

```python
lr=LinearRegression()
lr.fit(x_train, y_train)  # fir models with .fit method

svm1=SVR()
svm1.fit(x_train, y_train)

rf=RandomForestRegressor()
rf.fit(x_train, y_train)

gr=GradientBoostingRegressor()
gr.fit(x_train, y_train)
```

```python
y_predlr=lr.predict(x_test)
y_predsvm=svm1.predict(x_test)
y_predrf=rf.predict(x_test)
y_predgr=gr.predict(x_test)
```

# Evaluating  algorithems

```python
from sklearn import metrics
```

```python
scorelr=metrics.r2_score(y_test,y_predlr)
scoresvm=metrics.r2_score(y_test,y_predsvm)
scorerf=metrics.r2_score(y_test,y_predrf)
scoregr=metrics.r2_score(y_test,y_predgr)
```

```python
print (scorelr)
print (scoresvm)
print (scorerf)
print (scoregr)
```

```python
final_data=pd.DataFrame({"models":["LR", "SVR" , "RF" , "GR"] , "R2_Square":[scorelr,scoresvm,scorerf,scoregr]})
# making a data frame with r2 scores and models
```

```python
final_data
```

```python
sns.barplot(final_data["models"],final_data["R2_Square"])
```
